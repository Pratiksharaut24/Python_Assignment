```python
x = df.iloc[:,[3,4]].values
x
```

```python
c1 = KMeans(n_clusters=3, random_state=0, n_init="auto")
```

```python
c1.fit(x)
```

```python
y_labels = c1.fit_predict(x)
```

```python
y_labels
```

```python
df["Cluster"]=y_labels
```

```python
df.head(1)
```

```python
df["Cluster"].value_counts()
```

```python
wcss=[]
for i in range(1,10):
    c1=KMeans(n_clusters=i)
    c1.fit(x)
    wcss.append(c1.inertia_)       
```

```python
plt.plot(list(range(1,10)),wcss)
plt.xlabel("k")
plt.ylabel("wcss")
plt.show()
```
