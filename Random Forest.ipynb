```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
digits.data[:5]
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df["target"]=digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(["target"],axis="columns"),digits.target,test_size=0.20,random_state=45)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier                          ## Ensemble is use when you want to predict the multiple algorithms to predict the outcomes 
```

```python
mode=RandomForestClassifier()
```

```python
mode.fit(x_train,y_train)
```

```python
mode=RandomForestClassifier(n_estimators=40)
```

```python
mode.fit(x_train,y_train)
```

```python
mode.score(x_test,y_test)
```

```python
y_pred = mode.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cn = confusion_matrix(y_test,y_pred)
```

```python
cn
```

```python
import seaborn as sns
```

```python
sns.heatmap(cn,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
x = df[["Age", "EstimatedSalary"]].values
x
```

```python
y = df["Purchased"].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Classifier = RandomForestClassifier(n_estimators=100, criterion='gini', random_state=42)
```

```python
Classifier.fit(x_train,y_train)
```

```python
y_pred = Classifier.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cn = confusion_matrix(y_test,y_pred)
```

```python
cn
```

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(y_test,y_pred)
```

```python
score
```
