```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Mall_customers.csv")
df.head()
```

```python
from sklearn.metrics import silhouette_score
```

```python
x = df[['Annual Income (k$)', 'Spending Score (1-100)']].values	
x
```

```python
from sklearn.cluster import DBSCAN
```

```python
cluster = DBSCAN(eps=0.2,min_samples=3)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = StandardScaler().fit_transform(x)
```

```python
y_pred = cluster.fit_predict(x)
cluster
```

```python
y_pred  ## Noise point = -1(outlier)
```

```python
y_labels = cluster.labels_
y_labels
```

```python
x1 = df['Annual Income (k$)']
x2 = df['Spending Score (1-100)']
```

```python
plt.scatter(x1,x2,c=y_labels,cmap='rainbow',marker="*")
plt.legend(y_labels)
plt.show()
```

```python
df["cluster"]=y_labels
```

```python
df.head()
```

```python
silhouette_score(x,y_labels)
```
